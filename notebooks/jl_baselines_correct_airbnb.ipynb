```python
import pandas as pd
import pickle
from glob import glob
import numpy as np
```

```python
amenities_vars = ["Air conditioning", "Elevator", "Self check-in", "Pets allowed", "Private living room", "Backyard", "Pool"]
```

```python
for file in sorted(glob(f'/data/housing/data/intermediate/jl_pop_synth/airbnb_baselines/all_baselines_*.pickle')):
    city = file.split(".")[-2].split("_")[-1]
    with open(file, 'rb') as f:
        data = pickle.load(f)

    for k in ["df_nfvae", "df_vae", "df_nfvae95", "df_vae95"]:
        for var in [u for u in amenities_vars]:
            data[k][var] = (data[k][var] > data[k][var].quantile(1 - data[k.replace("nf", "").replace("vae", "real")][var].astype(np.float32).mean())) + 0.

        data[k][["accommodates","bathrooms","bedrooms","beds"]] = data[k][["accommodates","bathrooms","bedrooms","beds"]].astype(int) + 0.
    
    with open(file, 'wb') as f:
        pickle.dump(data, f, protocol=pickle.HIGHEST_PROTOCOL)
            
```
